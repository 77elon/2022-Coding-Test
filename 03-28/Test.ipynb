```python
result = 0
next_index = 0

#Test Case T + Cmp Case N
```

```python
# Found Max, Min 
# if stor[] is not Max, and Min <- stor[] >= min && stor[] < max buying
# if stor[] is Max <- selling
```

```python
def calc(ary):
    global result, next_index
    ary_max = np.max(ary)
    arg_max = np.argmax(ary)
    
    #for Slicing
    next_index = arg_max + 1
    temp = (ary_max - ary[:next_index])
    result += np.sum(temp)
```

```python
T = int(input())
```

```python
for i in range(T):
    result = 0
    next_index = 0

    N = int(input())
    stor = np.empty(0)
    stor = np.array(input().split(), dtype='int')
    
    while(len(stor) != next_index): # same as range N
        stor = stor[next_index:]
        #end num is biggest
        calc(stor)
```
